```python
# km rodados medio dos veiculos da base
# preco medio dos veiculos da base
# tabela de frequencia para cores do veiculo
# tabela de frequencia para ano/modelo
# preco medio e desvio padrao do km por carro, ordenar pela media
# preco medio e desvio padrao do km por ano/modelo, ordenar pelo ano/modelo
# preco medio e desvio padrao do km por carro x ano/modelo
# cortar km rodados em classe de 5 em 5 comecando em 0
# preco medio por classe de km rodado x carro

import pandas as pd
```

```python
df = pd.read_csv("./data/hb20_venda_webmotors_280314.txt", sep = "\t")
df
```

```python
# km rodados medio dos veiculos da base
# preco medio dos veiculos da base
df["km"].mean(), df["preco"].mean()
```

```python
# tabela de frequencia para cores do veiculo
df["cor"].value_counts().sum()
df2 = pd.DataFrame()
df2["frequencia"] = df["cor"].value_counts()
df2["pct"] = df2["frequencia"]/df["cor"].value_counts().sum()
df2
```

```python
# tabela de frequencia para ano/modelo
df["anomod"].value_counts().sum()
df2 = pd.DataFrame()
df2["frequencia"] = df["anomod"].value_counts()
df2["pct"] = df2["frequencia"]/df["anomod"].value_counts().sum()
df2
```

```python
# preco medio e desvio padrao do km por carro, ordenar pela media
df.groupby("carro").agg({"km":["mean", "std"]}).sort_values(by=("km", "mean"))
```
