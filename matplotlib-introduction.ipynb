# matplotlib example workflow

```python
# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # figsize =  width, height  ou largura, altura

# 3. Plot data
ax.plot(x,y) # add data to axes

# 4. Customize plot
ax.set(title="Simple plot",
      xlabel="x-axis",
      ylabel="y-axis")

# 5. Save and show
fig.savefig("img/sample-plot.png")
```

## Two options for subplots

```python
# Option one
fig, (ax, ax2) = plt.subplots(ncols=2)
```

```python
fig, (ax, ax2) = plt.subplots(nrows=2)
```

```python
fig, ((ax, ax2), (ax3,ax4)) = plt.subplots(nrows=2,
                                          ncols=2,
                                          figsize=(20,20))
#Plot do each different axes
ax.plot(x, x/2);
ax2.scatter(x, np.sin(x));
ax3.barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax3.set(title="Butter prices")
ax4.hist(x);
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                      ncols=2,
                      figsize=(15,15))
#Plot to each different index
ax[0,0].plot(x,x/2);
ax[0,1].scatter(x, np.sin(x))
ax[1,0].barh(list(nut_butter_prices.keys()),list(nut_butter_prices.values()))
ax[1,0].set(title="aeHOOOOO",
           xlabel="Preo")
ax[1,1].hist(x);
```

## Object oriented matplotlib with pandas
* When plotting something quick, ok to use pyplot method
* When plotting something advanced, use object oriented method

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50
```

```python
over_50.plot("age", "chol",kind = "scatter", c="target", figsize=(10,6))
```

```python
#OO method hybrid
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45,100])
ax.set(xlim=[45,100]) # esses 2 ultimos so iguais, mas prefiro o de baixo
```

```python
# OO method pure
fig, (ax, ax2) = plt.subplots(figsize=(10,10), nrows=2, sharex=True)

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])
# Customize
ax.set(title = "Heart Disease and cholesterol levels",
       ylabel= "Cholesterol")

# Add a legend
ax.legend(*scatter.legend_elements(), title= "Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");

# add new axes
scatter2 = ax2.scatter(x=over_50["age"],
           y=over_50["thalach"],
           c=over_50["target"])

ax2.set(xlabel= "Age",
       ylabel= "Max heart rate",
       title= "Max heart rate and heart disease")
ax2.legend(*scatter2.legend_elements(), title= "Target");
ax2.axhline(over_50["thalach"].mean(),
          linestyle="--");

# add tilte to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
plt.style.use("seaborn-whitegrid")
# OO method pure
fig, (ax, ax2) = plt.subplots(figsize=(10,10), nrows=2, sharex=True)

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="coolwarm") # this changes the color scheme
# Customize
ax.set(title = "Heart Disease and cholesterol levels",
       ylabel= "Cholesterol",
      xlim=[50,80],
      ylim=[100,500])

# Add a legend
ax.legend(*scatter.legend_elements(), title= "Target")

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle="--");

# add new axes
scatter2 = ax2.scatter(x=over_50["age"],
           y=over_50["thalach"],
           c=over_50["target"],
            cmap= "coolwarm")
ax2.set_xlim([50,80])
ax2.set_ylim([80,200]) # por alguma razao ele no deixa setar isso dentro do .scatter pro segundo grafico. pro primeiro vai de boa
ax2.set(xlabel= "Age",
       ylabel= "Max heart rate",
       title= "Max heart rate and heart disease")
ax2.legend(*scatter2.legend_elements(), title= "Target");
ax2.axhline(over_50["thalach"].mean(),
          linestyle="--");

# add tilte to figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
